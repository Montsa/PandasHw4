## Player Count

* Display the total number of players


```python
# Turn SN into set to remove duplicates, count elements in set to get total number of players
totalplayers = len(set(purchase_data['SN']))
print(f'total number of players: {totalplayers}')
```

## Purchasing Analysis (Total)

* Run basic calculations to obtain number of unique items, average price, etc.


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame


```python
# Turn Item ID into set to remove duplicates, count elements in set to get total unique items
uniqueitems = len(set(purchase_data['Item ID']))

# Use agg function to find average price, number of purchase, and total revenue
avgprice = purchase_data['Price'].agg(np.mean)
numofpurchase = purchase_data['Purchase ID'].agg('count')
totalrevenue = purchase_data['Price'].agg('sum')

#create data frame to hold values
#create list for titles and list of list for values
summary_titles = ['Number of Unique Items', 'Average Price','Number of Purchases','Total Revenue']
summary_values = [[x] for x in [uniqueitems,avgprice,numofpurchase,totalrevenue]]

#convert lists into dictionary
summary_list = list(zip(summary_titles,summary_values))
summary_dict = dict(summary_list)

#convert dictionary to DataFram
df_summary = pd.DataFrame(summary_dict)

#Format DataFrame columns
df_summary['Average Price'] = df_summary['Average Price'].map("${:,.2f}".format)
df_summary['Total Revenue'] = df_summary['Total Revenue'].map("${:,.2f}".format)

#Display DataFrame
df_summary
```

## Gender Demographics

* Percentage and Count of Male Players


* Percentage and Count of Female Players


* Percentage and Count of Other / Non-Disclosed




```python
#remove duplicate player occurences
df_nodupplayers = purchase_data.drop_duplicates(subset='SN')

#create a DataFRame
df_GenderDemo = pd.DataFrame(df_nodupplayers['Gender'])

#add columns for Total Count and Percentage of Players
df_GenderDemo = pd.DataFrame({'Total Count':df_GenderDemo.groupby(['Gender']).Gender.agg('count')})
df_GenderDemo['Percentage of Players'] = [(x/totalplayers)*100 for x in df_GenderDemo['Total Count']]

#Format DataFrame and display
df_GenderDemo['Percentage of Players'] = df_GenderDemo['Percentage of Players'].map("{:,.2f}".format)
df_GenderDemo
```

```python
# Goal is to bin price 

def bin_agecount(age):
    if age < 10:
        return "<10"
    elif age <= 14:
        return "10-14"
    elif age <= 19:
        return "15-19"
    elif age <= 24:
        return "20-24"
    elif age <= 29:
        return "25-29"
    elif age <= 34:
        return "30-34"
    elif age <= 39:
        return "35-39"
    else:
        return "40+"
    
bin_agecount(23)
```

```python
# Create DataFrame of Unique players
df_AgeDemo = pd.DataFrame(df_nodupplayers)

#Add new column 'Age Bin' that bins players ages
df_AgeDemo['Age Bin'] = df_AgeDemo['Age'].map(bin_agecount)

#Group players by 'Age_Bin' and count number of occurences in each bin
df_AgeDemo1 = pd.DataFrame({'Total Count':df_AgeDemo.groupby(['Age Bin']).Price.agg('count')})

#Create new column, ' Percentage of Players', to calulcate the number of Players in each bin
df_AgeDemo1['Percentage of Players'] = [(x/totalplayers)*100 for x in df_AgeDemo1['Total Count']]

#Display DataFrame
df_AgeDemo1
```

## Top Spenders
